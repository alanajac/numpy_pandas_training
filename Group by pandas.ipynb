```python
#Very similar to SQL.
#Using groupby() creates a lazy
import numpy as np
import pandas as pd
df=pd.read_csv('mpg.csv')
df
#How to check which variables would be categorized
df['model_year'].nunique()
```

```python
#When I use groupby, it creates an object.So we have to ask for another method to give us some information
#This way we can group by and get the mean of a specific column. In this case, miles per galon. 
df.groupby('model_year').mean()['mpg']
```

```python
#Lets pass now a list.
df.groupby(['model_year','cylinders']).mean()
#We can also check the columns we have:
df.groupby(['model_year','cylinders']).mean().columns
```

```python
#And we can also check the index
df.groupby(['model_year','cylinders']).mean().index
#And we can also get the mean of a specific column:
df.groupby(['model_year','cylinders']).mean(['mpg'])
#But always the best method is describe!
```

```python
#The best method is still describe, to have a general view of the table
df.groupby(['model_year']).describe()
```

```python
#We want a multi-index data
year_cyl = df.groupby(['model_year','cylinders']).mean()
```

```python
year_cyl
#we can also use the index.levels, to check the levels of the index
year_cyl.index.levels
#So it shows a list of the different indexes/groups you have by the multi-indexes choice
```

```python
#we can also check only one column, by using:
#THis because we have a index.
year_cyl.loc[70]
```

```python
#we can also ask from a list of one of the indexes
year_cyl.loc[[70,82]]
```

```python
#Or a combination of the two indexes:
year_cyl.loc[(70,4)]
#And then we get the exact row which attends both 
```

```python
#Another method!It does a cross-section!
#year_cyl.xs(key=value,level='category')
year_cyl.xs(key=70,level='model_year')
#We can also pass a list of keys by using loc
year_cyl.loc[[70,80]]
#And if we want the instance of a specific value for all the groups of a given index.
#In this case its not showing the cylinders column because it is the one chosen, with the same value.
#What we can do its to assign one variable to the result.
year_cyl.xs(key=4,level='cylinders')
#such as:
four_cyl = year_cyl.xs(key=4,level='cylinders')
#But lets now filter one of the indexes.
#Lets choose only the ones without the rows with cylinders = 4
df[df['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean()
```

```python
#we can also swap the levels
year_cyl.swaplevel()
#It will just change the order of the indexes used.
```

```python
#And of course, we can sort the levels the way we want!
#Same method as before, just assigning the level we want.
year_cyl.sort_index(level='model_year',ascending=False)
#Important: Always better to sort by the outer level.
```
