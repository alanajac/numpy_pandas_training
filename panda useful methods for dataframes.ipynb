```python
#Lets do this!
#First for single columns and then to multiple columns
import pandas as pd, numpy as np
df=pd.read_csv('tips.csv')
df.info
#Lets do an example. Lets get the last 4 numbers of the credit card.
#But we see that the credit card number is registered as an int, and not as an integer.
#So its difficult to do it. :D We need to first convert it for a string and then get the last 4 numbers of
#the credit card. Lets define a function for that:
def last_four(cc_number):
    return str(cc_number)[-4:]
#Using apply we can use a function to be applied to the data frame!

df['last_four']=df['CC Number'].apply(last_four)
df
print(df['total_bill'].mean())
def yelp(price):
    if price<10:
        return '$'
    elif price>=10 and price <30:
        return '$$'
    else: 
        return '$$$'
df
df['expenses']=df['total_bill'].apply(yelp)
#apply is a single value function. Which is applied to every row. :)

df
#NOW MULTIPLE COLUMNS FUNCTIONS!
#First lets use lambda expressions
#the function
def multiply_by_two(num):
    return num*2
#wen can represented by
lambda num:num*2
#So we can do something like:
df['total_bill'].apply(lambda num:num*2)
df
#lets create a new function
def quality(total_bill,tip):
    if tip/total_bill>0.25:
        return 'generous'
    else: return 'other'
    
#testing...
quality(17.01,1.99)
#Now lets apply this function using the apply method
#Check the typos. The first argument of function applied is quality, and the second is axis=1 -> columns
df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
#we can now add the column to the data_frame
df['quality']=df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
df
```

```python
#We can do it in a simpler way:
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
#np.vectorize will help us with gain/efficiency
#Lets check how fast is the difference!
import timeit
```

```python
#Lets try another methods. We used timeit
#Lets read again the same table:
#we can use apply to do the transformation of the integer to string.
#Lets check which one is the fastest
setup='''
import numpy as np
import pandas as pd
df=pd.read_csv('tips.csv')
'''

com1='''
df['CC_Number']=df['CC Number'].apply(str)
'''
com2='''
df['CC_Number']=df['CC Number'].map(str)
'''
print(timeit.timeit(setup=setup,stmt=com1,number=1000 ))
print(timeit.timeit(setup=setup,stmt=com2,number=1000 ))
#map is faster to do the calculation than apply!:D


```
